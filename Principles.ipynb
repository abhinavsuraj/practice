### Assumptions of Linear Regressions <a class="anchor" id="assumptions_of_LR"></a>
[Source](https://www.statology.org/linear-regression-assumptions/)

There are four assumptions associated with a linear regression model:

- ***Linearity***: The relationship between X and the mean of Y is linear.
- ***Homoscedasticity***: The variance of residual is the same for any value of X.
- ***Independence***: Observations are independent of each other.
- ***Normality***: For any fixed value of X, Y is normally distributed.




### L1 Vs L2 Regularization <a class="anchor" id="l1_vs_l2"></a>

Overfitting is a phenomenon that occurs when a machine learning or statistics model is tailored to a particular dataset and is unable to generalise to other datasets. Regularisation is a process of introducing additional information in order to prevent overfitting

***The key difference between these two is the penalty term.***
- **L1 Regularization** or **Lasso**(Least Absolute Shrinkage and Selection Operator):  adds “absolute value of magnitude” of coefficient as penalty term to the loss function.

***So, Lasso works well for feature selection in case we have a huge number of features.***




- **L2 Regularization** or **Ridge**: adds “squared magnitude” of coefficient as penalty term to the loss function


### Variance Bias Tradeoff <a class="anchor" id="Variance_Bias"></a>

- **Bias** Tendency to learn the wrong pattern
    - ◦ **Low bias classifier** = makes less assumptions about the form of the relationship between x and y
        - Suggests less assumptions about the form of the target function.
        - <b>Example</b>: Decision Trees, k-Nearest Neighbors and Support Vector Machines
    - ◦ **High bias classifier** = makes many assumptions about the form of the relationship between x and y
        - Suggests more assumptions about the form of the target function.
        - <b>Example</b>: Linear Regression, Linear Discriminant Analysis and Logistic Regression.

- **Variance** Tendency to learn a random pattern irrespective of what the real pattern is
    - ◦ **Low variance classifier**: small changes in the training data will not affect the estimated relationship much
        - Suggests small changes to the estimate of the target function with changes to the training dataset.
        - <b>Example</b> Linear Regression, Linear Discriminant Analysis and Logistic Regression
    - ◦ **High variance classifier**: small changes in the training data will affect the estimated relationship significantly
        - Suggests large changes to the estimate of the target function with changes to the training dataset.
        - Generally nonparametric machine learning algorithms that have a lot of  exibility have a high variance.
        - <b>Example</b>: Decision trees have a high variance, that is even higher if the trees are not pruned before use
            - Decision trees, k-Nearest Neighbors and Support Vector Machines.
            
- Parametric or linear machine learning algorithms often have a high bias but a low variance.
- Nonparametric or nonlinear machine learning algorithms often have a low bias but a high variance.

-  The k-nearest neighbors algorithm has low bias and high variance, but the trade-off can be changed by increasing the value of k which increases the number of neighbors that contribute to the prediction and in turn increases the bias of the model.
-  The support vector machine algorithm has low bias and high variance, but the trade-off can be changed by increasing the C parameter that infuences the number of violations of the margin allowed in the training data which increases the bias but decreases the variance.

### Overfitting Vs Underfitting <a class="anchor" id="overfitting_vs_underfitting"></a>

- **Overfitting** Tendency to learn the wrong pattern
    - ◦ Overfitting refers to a model that models the training data too well. It happens when a model **learns the detail and noise in the training data** to the extent that it negatively impacts the performance on the model on new data. This means that the noise or random  fluctuations in the training data is picked up and learned as concepts by the model. The problem is that these concepts do not apply to new data and negatively impact the models ability to generalize.
    - ◦ **Can be avoided**
        - Use a resampling technique to estimate model accuracy.(**K-Fold**)
        - Hold back a validation dataset.

- **Underfitting** Underfitting refers to a model that can neither model the training data nor generalize to new data. An underfit machine learning model is not a suitable model and will be obvious as it will have poor performance on the training data.

```python
def bubble_sort(arr):
    def swap(i,j):
        arr[i],arr[j]=arr[j],arr[i]
    n = len(arr)
    swap_flag = True
    x=-1
    while swap_flag:
        swap_flag = False
        x= x+1
        for i in range(1,n-x):
            if arr[i-1] > arr [i]:
                swap(i-1,i)
                swap_flag = True
    
    return arr
```
