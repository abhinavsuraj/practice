## Power of Hypothesis Testing <a class="anchor" id="about"></a>

### When and where it's used?
- Statistical Power is used in binary Hypothesis Test

### Definition
- It is the probability that a test rejects the null hypothesis correctly when the alternative Hypothesis is true (<b>1 - $\beta$</b>)

- It is the likelihood that a test will detect an effect, when an effect is present

### When Value Changed
- The higher the statistical power, the better the test is

### Application / Example
- It is commonly used in Experiment Design
- To calculate the minimum sample size so that one can reasonably detect an effect

## Type I Error <a class="anchor" id="type1"></a>
- It is also known as <b>False Positive</b>

### When and where it's used?
- It is used to categorize errors in a binary Hypothesis Test

### Definition
- It occurs when we Mistakely reject  <b>true</b> null hypothesis
- It means that we conclude our findings are significant, when in fact it had occured by chance

### When Value Changed
- Larger the value -> the less reliable a test is

### Application / Example
- It is commonly used in A/B Testing
- To show that we observe differences between two groups, but in reality there is no difference

## $\alpha$ Value  <a class="anchor" id="alpha"></a>
### The probability of making a Type I error is $\alpha$ (alpha)
- SIgnificance Level

## Type II Error <a class="anchor" id="type2"></a>
- It is also known as <b>False Negative</b>

### When and where it's used?
- It is used to categorize errors in a binary Hypothesis Test

### Definition
- It occurs when we fail to reject  <b>false</b> null hypothesis
- It means that we conclude our findings are not significant, when in fact it there is a significant effect

### When Value Changed
- Larger the value -> the less reliable a test is
- We want to minimize this error

### Application / Example
- It is commonly used in A/B Testing
- To show that we don't observe differences between two groups, but in reality there is significant difference

## Confidence Interval <a class="anchor" id="confidence_interval"></a>

### When and where it's used?
- Want to know how variable a sample result might be
- It is for estimating true value, but we never know what the true value is
- The purpose of having samples and observations is to estimate the true value


### Definition
- CI is range of numbers 
- It tells how often it contains true value
- The Probability of CI covering the true value is <b> Confidence Level </b>
- Commonly used Confidence Level is 95%

### When Value Changed
- Wider the interval -> the more uncertain we are about the sample result
- So more confident we want to be, the less data we have -> Wider the CI to be able to capture the true value
- Higher Confidence level -> Wider CI

### Common Misconception
- CI answers: What is the Probability true value is in the CI <b> Completely wrong </b>
- True value is a variable & CI is deterministic <b> Completely wrong </b>
- True Value: Determined by nature, but is unknown to us
- CI:
    - Variable
    - Estimated from samples and Confidence Level we set
- For a particular CI:
    - The true value is either 100% in it or not
    - Confidence Level: How likely CI(Estimated from samples) covers teh true value
### In simple terms
- CI measures a level of Uncertainity when we try to estimate the true value

#### Example
- Measure average weight of men in US
    - Randomly measure weight of 30 men and get a CI
    - CI will likely cover true weight of all men in US
    - But how likely
        - If we repeate the process over and over, we expect the CI we construct will cover true value 95% of the time
       

## Confidence Level <a class="anchor" id="confidence_level"></a>
https://statisticsbyjim.com/hypothesis-testing/confidence-intervals-compare-means/
### When and where it's used?


### Definition
- Different random samples drawn from the same population are liable to produce slightly different intervals. If you draw many random samples and calculate a confidence interval for each sample, a specific proportion of the intervals contain the population parameter. That percentage is the confidence level.
- The Probability of CI covering the true value is <b> Confidence Level </b>
- Commonly used Confidence Level is 95%

### Example
- A 95% confidence level suggests that if you draw 20 random samples from the same population, youd expect 19 of the confidence intervals to include the population value.

### When Value Changed
- Lowering the confidence level ->  will create a narrower CI, which looks good. However, it comes at the expense of dramatically increasing the likelihood that the CI wont contain the correct population value
- Higher Confidence level -> Wider CI


### Confidence Intervals and P Values Always Agree on Statistical Significance
- If you want to determine whether your test results are statistically significant, you can use either P values with significance levels or confidence intervals. These two approaches always agree
- Confidence level = 1  Significance level (alpha)

### Both of the following conditions represent a hypothesis test with statistically significant results:

- The P value is <b>smaller than</b> the significance level.
- The confidence interval <b>excludes</b> the null hypothesis value.

## P Value <a class="anchor" id="pvalue"></a>

### When and where it's used?
- It is commonly used in Hypothesis Testing to connect the dots between Observation and Conclusion

### Definition
- It is a conditional probability
- Measures Pr(results as extreme as observed results / null hypothesis is true)
- [] Probability of obtaining a sample <b>more extreme</b> than the ones observed in your data assuming H0 is true

### When Value Changed
- Low p value -> the less support of null hypothesis
- We Often use 0.05 in practice as a cutoff value (<b>Significance Level </b>)
- p value < 0.05: denotes strong evidence against null hupothesis -> reject null hypothesis
- p value > 0.05: denotes weak evidence against null hupothesis -> fail to reject null hypothesis or cannot be rejected

### Application / Example
- It is commonly used in A/B Testing
- When we have treatment and control groups, and we want to test if a metric is different in two groups
- Suppose we obtained an experiment and obtained measurements from two groups
    - Smaller the p value: more we are convinced there is a difference between 2 groups

#### Common Mistake
- p value = Pr(there's a difference between groups/ observation) <b> Completely wrong </b>
                - OR
- p value = Pr(null hypothesis is true/ observation) <b> Completely wrong </b>


- p value = Pr(there's a difference/ null hypothesis is true) <b> Correct Interpretation </b>
    - Given null hypothesis is true, the probability of observing differences is p values

### In Other Words
- Question: Average weight of US men is same as fixed value(195 lbs)
- p value connects what data we observed and what conclusion we could draw
- It tells us assuming the true average weight = 195 lbs, how likely we observe data?
- Small p-value ex: 0.05 means, Assume avg. weight is 195 lbs, chance that we obseve the data is very low or data we observe is very extreme. So we believe the true value should not be 195 lbs

### Application / Example
It is commonly used in Experiment Design
To calculate the minimum sample size so that one can reasonably detect an effect

### When we do hypothesis testing, we are always asking... DOES the evidence that we collected make our null hypothesis look ridiculous or not?
- What p-value <b>DOES IS</b> provide an answer to that question. It tells you essentially wether your evidenvce make your null hypothesis look ridiculous or not.
- What it <b>IS</b> is the probability of getting something at least as bad as what we got in that world that we constructed.

## Significance Level <a class="anchor" id="significance"></a>
- It defines the strength of evidence in probabilistic terms
- It is equal to  $\alpha$ (Probability of False Positive) for a test
- It presents the probability that tests will produce statistically significant results when the null hypothesis is correct.

### When and where it's used?
- In the testing process, you use significance levels and p-values to determine whether the test results are statistically significant.

## Definition
- <b>The significance level is a measure of how strong the sample evidence must be before determining the results are statistically significant</b>
- It defines how strongly the sample evidence must contradict the null hypothesis before you can reject the null hypothesis for the entire population. 


### When Value Changed
- Increased the value (from 0.05 to 0.1) -> Because this change decreases the amount of required evidence, it makes your test more sensitive to detecting differences, but it also increases the chance of a false positive from 5% to 10%.
- Decreased the value (from 0.05 to 0.01) -> Because this change increases the amount of required evidence, it makes your test less sensitive to detecting differences, but it decreases the chance of a false positive from 5% to 1%.
- Its all about the tradeoff between sensitivity and false positives

### Application / Example
- It is commonly used in A/B Testing or Hypothesis testing
